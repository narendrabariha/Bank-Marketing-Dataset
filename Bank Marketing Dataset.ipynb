## Histogram plot

```python
plt.figure(figsize=(6,3))
# Histogram plot
sns.histplot(df["age"],bins=30) 

plt.title("Histogram plot for age",y=1.02)
plt.xlabel("age")
plt.ylabel("Frequency")
plt.show()
```

## Distribution plot

```python
plt.figure(figsize=(6,3))
# Distribution plot
sns.displot(df["age"],kind="kde") 
plt.title("Density plot for age",y=1.02)
plt.xlabel("age")
plt.ylabel("Probability")
plt.show()
```

## Box plot

```python
plt.figure(figsize=(6,3))
# Box plot
sns.boxplot(df["age"],orient="h") 

plt.title("Box plot for age",y=1.02)
plt.xlabel("age")
# plt.ylabel("")
plt.show()
```

## Violin Plot

```python
plt.figure(figsize=(8,5))
# Violin Plot
sns.violinplot(df["age"],orient="h",color="chartreuse") 
sns.swarmplot( df["age"],size=0.3,orient="h")

plt.title("Violin plot for age",y=1.02)
plt.xlabel("age")
# plt.ylabel("")
plt.show()
```

```python
df["age"].quantile(0.10),df["age"].quantile(0.25),df["age"].quantile(0.50),df["age"].quantile(0.75),df["age"].quantile(0.90)
```

**age** column has a symmetrical distribution with skew 0.86 which is between 0.5 and 1 .

- Density curve for age is centred around mean 41.23 with standard deviation of 11.91
- Range of age value is from minimum 18 and maximum 95
- Most of the people are failling under age group of 33 - 58
- 10% of people are of age 28.0 and below.
- 25% of people are of age 32.0 and below.
- 50% of people are of age 39.0 and below.
- 75% of people are of age 49.0 and below.
- 90% of people are of age 58.0 and below.
- age column has few extreme values toward higher age group 73 and above.

age column could be a good predictor for balance and its relation with other columns should be explored further.
